## Import Libaries

## Organizing

```
SQLMoviSer.release_date=pd.to_datetime(Moviesdata2.release_date)
```

```
SQLMoviSer
```

```
SQLMoviSer.revenue=Moviesdata2.revenue/1000000
SQLMoviSer.budget=Moviesdata2.budget/1000000
```

```
SQLMoviSer
```

```
ratings=Moviesdata2[['id',"vote_count","vote_average"]].copy()
ratings
```

```
moviesgenres=pd.json_normalize(data=data,record_path='genres',meta='id',record_prefix='genre_')
moviesgenres
```

```
companyproduction=pd.json_normalize(data=data,record_path='production_companies',meta="id",record_prefix='com[')
companyproduction
```

```
con=sq3.connect('movies.db')
```

```
con
```

```
SQLMoviSer.to_sql("Movies",con,index=False)
```

```
ratings.to_sql("Votes",con,index=False)
```

```
moviesgenres.to_sql("Genres",con,index=False)
```

```
companyproduction.to_sql("Prod",con,index=False)
```

```
con.execute("Select * FROM sqlite_master").fetchall()
```

```
con.execute("Select name FROM sqlite_master WHERE type='table' ORDER BY name").fetchall()
```

```
con.close()
```

###Loading Data from SQL 

```
import sqlite3 as sq3
import pandas as pd
```

```
con=sq3.connect("movies.db")
```

```
pd.read_sql("SELECT * FROM Movies",con)
```

```
pd.read_sql("SELECT * FROM Movies",con).describe()
```

```
pd.read_sql("SELECT * FROM Movies",con).mean()
```

```
Dataframe=pd.read_sql("SELECT * FROM Movies",con,index_col="id",parse_dates="release_date")
Dataframe
```

```
Productions2=pd.read_sql("SELECT * FROM Prod",con,index_col="com")
Productions2
```

Diffent Queries

```
con=sq3.connect("movies.db")
```

```
pd.read_sql("SELECT *  FROM Movies JOIN Votes ON Movies.id=Votes.id",con)
```

```
pd.read_sql("SELECT Movies.id,Movies.title,Votes.vote_average FROM Movies JOIN Votes ON Movies.id=Votes.id WHERE Votes.vote_average>8 ORDER BY Movies.revenue ASC",con,index_col='id')
```
